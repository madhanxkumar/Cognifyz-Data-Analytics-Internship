```python
import pandas as pd

df = pd.read_csv("Downloads/Dataset .csv")

df.head()
```

```python
rating_distribution = df['Aggregate rating'].value_counts().sort_index()

print("Rating Distribution:")
print(rating_distribution)
```

```python
most_common_rating = df['Aggregate rating'].mode()[0]

print("Most Common Rating:", most_common_rating)
```

```python
average_votes = df['Votes'].mean()

print("Average Number of Votes:", round(average_votes, 2))
```

```python
print(df['Aggregate rating'].describe())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.hist(
    df['Aggregate rating'],
    bins=10,
    color='#4ECDC4',
    edgecolor='black'
)

plt.title('Distribution of Aggregate Ratings')
plt.xlabel('Aggregate Rating')
plt.ylabel('Number of Restaurants')

plt.show()
```

```python
rating_votes = df.groupby('Aggregate rating')['Votes'].mean()

rating_votes.head(10)
```

```python
rating_votes.head(10).plot(
    kind='bar',
    color='#6C5CE7'
)

plt.title('Average Votes by Rating')
plt.xlabel('Aggregate Rating')
plt.ylabel('Average Votes')

plt.show()
```
