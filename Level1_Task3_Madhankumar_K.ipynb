```python
import pandas as pd

df = pd.read_csv("Downloads/Dataset .csv")

df.head()
```

```python
price_range_count = df['Price range'].value_counts().sort_index()

print("Price Range Distribution:")
print(price_range_count)
```

```python
most_common_price_range = price_range_count.idxmax()
highest_count = price_range_count.max()

print("Most Common Price Range:", most_common_price_range)
print("Number of Restaurants:", highest_count)
```

```python
price_percentage = (price_range_count / len(df)) * 100
print("Percentage Distribution:")
print(round(price_percentage, 2))
```

```python
result = pd.DataFrame({
    'Count': price_range_count,
    'Percentage': round(price_percentage, 2)
})

result
```

```python
import matplotlib.pyplot as plt

ax = price_range_count.plot(
    kind='bar',
    color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#6C5CE7']
)
plt.title('Price Range Distribution', fontsize=14, fontweight='bold')
plt.xlabel('Price Range')
plt.ylabel('Number of Restaurants')

for bar in ax.patches:
    ax.annotate(
        str(int(bar.get_height())),
        (bar.get_x() + bar.get_width()/2, bar.get_height()),
        ha='center',
        va='bottom'
    )

plt.xticks(rotation=0)
plt.show()
```
