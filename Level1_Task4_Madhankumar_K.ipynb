```python
import pandas as pd

df = pd.read_csv("Downloads/Dataset .csv")

df.head()
```

```python
online_delivery_count = df['Has Online delivery'].value_counts()

print("Online Delivery Distribution:")
print(online_delivery_count)
```

```python
online_delivery_percentage = (online_delivery_count / len(df)) * 100

print("Percentage Distribution:")
print(round(online_delivery_percentage, 2))
```

```python
delivery_rating = df.groupby('Has Online delivery')['Aggregate rating'].mean()

print("Average Rating Comparison:")
print(round(delivery_rating, 2))
```

```python
result = pd.DataFrame({
    'Restaurant Count': online_delivery_count,
    'Percentage': round(online_delivery_percentage, 2)
})

result
```

```python
import matplotlib.pyplot as plt

online_delivery_count.plot(
    kind='pie',
    autopct='%1.1f%%',
    colors=['#FF6B6B', '#4ECDC4'],
    startangle=90
)

plt.title('Online Delivery Availability')
plt.ylabel('')
plt.show()
```

```python
delivery_rating.plot(
    kind='bar',
    color=['#45B7D1', '#6C5CE7']
)

plt.title('Average Rating by Online Delivery Availability')
plt.xlabel('Online Delivery')
plt.ylabel('Average Rating')
plt.xticks(rotation=0)

for i in plt.gca().patches:
    plt.gca().annotate(
        f'{i.get_height():.2f}',
        (i.get_x() + i.get_width()/2, i.get_height()),
        ha='center',
        va='bottom'
    )

plt.show()
```
