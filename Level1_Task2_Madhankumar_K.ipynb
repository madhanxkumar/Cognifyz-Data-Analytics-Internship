```python
import pandas as pd

df = pd.read_csv("Downloads/Dataset .csv")

df.head()
```

```python
city_count = df['City'].value_counts()

print("Top 10 Cities by Number of Restaurants:")
print(city_count.head(10))
```

```python
highest_city = city_count.idxmax()
highest_count = city_count.max()

print("City with Highest Number of Restaurants:")
print(highest_city)
print("Number of Restaurants:", highest_count)
```

```python
city_rating = df.groupby('City')['Aggregate rating'].mean()

print("Top 10 Cities by Average Rating:")
print(city_rating.sort_values(ascending=False).head(10))
```

```python
highest_rated_city = city_rating.idxmax()
highest_rating = city_rating.max()

print("City with Highest Average Rating:")
print(highest_rated_city)
print("Average Rating:", round(highest_rating, 2))
```

```python
import matplotlib.pyplot as plt

top10_cities = city_count.head(10)

top10_cities.plot(
    kind='bar',
    color=['#FF6B6B','#4ECDC4','#45B7D1','#96CEB4',
           '#FFEAA7','#DDA0DD','#F8B195','#6C5CE7',
           '#00B894','#E17055']
)

plt.title('Top 10 Cities by Number of Restaurants')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45)
plt.show()
```
