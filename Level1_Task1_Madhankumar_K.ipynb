```python
import pandas as pd
```

```python
df = pd.read_csv("Downloads/Dataset .csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
top_cuisines = df['Cuisines'].value_counts().head(3)

print("Top 3 Most Common Cuisines:")
print(top_cuisines)
```

```python
total_restaurants = len(df)

cuisine_percentage = (top_cuisines / total_restaurants) * 100

print("Percentage of Restaurants Serving Top Cuisines:")
print(cuisine_percentage)
```

```python
import matplotlib.pyplot as plt

top_cuisines.plot(
    kind='bar',
    color=['#FF6B6B', '#4ECDC4', '#45B7D1']
)

plt.title('Top 3 Most Common Cuisines', fontsize=14, fontweight='bold')
plt.xlabel('Cuisine')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=30)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
result = pd.DataFrame({
    'Count': top_cuisines,
    'Percentage': cuisine_percentage
})

result
```
