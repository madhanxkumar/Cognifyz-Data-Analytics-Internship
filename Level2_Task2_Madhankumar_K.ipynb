```python
import pandas as pd

df = pd.read_csv("Downloads/Dataset .csv")

df.head()
```

```python
top_cuisine_combinations = df['Cuisines'].value_counts().head(10)

print("Top 10 Cuisine Combinations:")
print(top_cuisine_combinations)
```

```python
cuisine_ratings = df.groupby('Cuisines')['Aggregate rating'].mean()

top_rated_cuisines = cuisine_ratings.sort_values(ascending=False).head(10)

print("Top 10 Highest Rated Cuisine Combinations:")
print(top_rated_cuisines)
```

```python
result = pd.DataFrame({
    'Restaurant Count': top_cuisine_combinations
})

result
```

```python
import matplotlib.pyplot as plt

top_cuisine_combinations.plot(
    kind='bar',
    color=['#FF6B6B','#4ECDC4','#45B7D1','#96CEB4',
           '#FFEAA7','#DDA0DD','#F8B195','#6C5CE7',
           '#00B894','#E17055']
)

plt.title('Top 10 Cuisine Combinations')
plt.xlabel('Cuisine Combination')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=75)

plt.show()
```

```python
highest_rated_cuisine = cuisine_ratings.idxmax()
highest_rating = cuisine_ratings.max()

print("Highest Rated Cuisine Combination:")
print(highest_rated_cuisine)
print("Average Rating:", round(highest_rating, 2))
```
